```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter
import os
```

```python
dir, file = os.path.split(r"I:\20200810_CameraTest\few\result.csv")
```

```python
def draw_himstogram(file_dir):
    _bin = np.arange(0,0.1,0.005)  
    dir, file = os.path.split(file_dir)
    data = pd.read_csv(file_dir, index_col = 0)
    flatten_data = data.to_numpy(). flatten() 
    flattend_data_diff = flatten_data-0.21
    flattend_data_diff = np.abs(flatten_data-0.21)
    plt.hist(flattend_data_diff, bins = _bin, weights = np.ones(len(flattend_data_diff))/len(flattend_data_diff))
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    print(np.average(flattend_data_diff))
    plt.title("["+file+"] "+"error (cm)")
    plt.show()
```

```python
draw_himstogram(r"C:\Users\ZAIO\dev\python_aruco_marker_visualization\camera_aruco_distance_test\1m.csv")
draw_himstogram(r"C:\Users\ZAIO\dev\python_aruco_marker_visualization\camera_aruco_distance_test\3m.csv")
draw_himstogram(r"C:\Users\ZAIO\dev\python_aruco_marker_visualization\camera_aruco_distance_test\5m.csv")
draw_himstogram(r"C:\Users\ZAIO\dev\python_aruco_marker_visualization\camera_aruco_distance_test\floor.csv")
```

```python
## boxplot
data_dir = r"C:\Users\ZAIO\dev\python_aruco_marker_visualization\camera_aruco_distance_test"
exp_list= ["1m","3m","5m"]
data_dic = {}

for file_name in exp_list:
    file_dir = os.path.join(data_dir,file_name+".csv")
    data = pd.read_csv(file_dir, index_col = 0)
    flatten_data = data.to_numpy().flatten()
    flatten_data = (flatten_data-0.21)*100
    # flatten_data = np.abs(flatten_data)
    data_dic[file_name] = flatten_data
```

```python
data = [data_dic["1m"], data_dic["3m"], data_dic["5m"]]
fig7, ax7 = plt.subplots()
ax7.set_title('Multiple Samples with Different sizes')
ax7.boxplot(data,showcaps =False)

```
